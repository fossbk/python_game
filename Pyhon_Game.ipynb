# Trò chơi "Đoán số" version 1
Trò chơi Đoán số là một trò chơi thú vị và đơn giản, nơi người chơi phải đoán một số mà máy tính đã chọn ngẫu nhiên trong một khoảng cho trước. Mỗi khi người chơi đoán một số, hệ thống sẽ cung cấp gợi ý về việc số đoán là "quá cao" hay "quá thấp" so với số máy tính đã chọn, giúp người chơi tiến dần đến đáp án chính xác.

# Mục tiêu:
Đoán đúng số mà máy tính chọn trong thời gian nhanh nhất và với ít lần đoán nhất.

# Hướng dẫn cách chơi:
Khởi động trò chơi:

Máy tính sẽ chọn một số ngẫu nhiên trong khoảng từ 1 đến 100.

# Lượt chơi:

Người chơi nhập một số vào (có thể là số nguyên trong khoảng 1-100).

Máy tính sẽ kiểm tra số đoán và thông báo kết quả:

Nếu số đoán quá thấp: Hệ thống sẽ gợi ý người chơi nhập số lớn hơn.

Nếu số đoán quá cao: Hệ thống sẽ gợi ý người chơi nhập số nhỏ hơn.

Nếu người chơi đoán đúng, hệ thống sẽ chúc mừng và kết thúc trò chơi.

# Kết thúc trò chơi:

Trò chơi kết thúc khi người chơi đoán đúng số máy tính đã chọn.

Sau khi đoán đúng, trò chơi sẽ thông báo số lượt đoán mà người chơi đã thực hiện.



```python
import random

def guess_the_number():
    # Máy tính chọn số ngẫu nhiên trong phạm vi 1 đến 100
    number_to_guess = random.randint(1, 100)
    attempts = 0

    print("Chào mừng bạn đến với trò chơi Đoán Số!")
    print("Tôi đã chọn một số ngẫu nhiên giữa 1 và 100. Hãy đoán xem đó là số nào!")

    while True:
        # Người chơi nhập số dự đoán
        player_guess = int(input("Nhập số bạn đoán: "))
        attempts += 1

        # Kiểm tra kết quả
        if player_guess < number_to_guess:
            print("Số bạn đoán nhỏ hơn số tôi chọn. Hãy thử lại!")
        elif player_guess > number_to_guess:
            print("Số bạn đoán lớn hơn số tôi chọn. Hãy thử lại!")
        else:
            print(f"Chúc mừng bạn! Bạn đã đoán đúng số {number_to_guess} sau {attempts} lần thử!")
            break
# Gọi hàm để chơi trò chơi
guess_the_number()
```

Cho phép người chơi nhập vào phạm vi n1 và n2 cho trò chơi "Đoán số" version 2

```python
import random

def guess_the_number():
    # Nhập phạm vi số mà người chơi muốn
    n1 = int(input("Nhập giá trị nhỏ hơn trong phạm vi đoán: "))
    n2 = int(input("Nhập giá trị lớn hơn trong phạm vi đoán: "))

    # Kiểm tra để đảm bảo n1 nhỏ hơn n2
    if n1 >= n2:
        print("Phạm vi không hợp lệ. Giá trị nhỏ nhất phải nhỏ hơn giá trị lớn nhất.")
        return

    # Máy tính chọn số ngẫu nhiên trong phạm vi n1 đến n2
    number_to_guess = random.randint(n1, n2)
    attempts = 0

    print(f"Chào mừng bạn đến với trò chơi Đoán Số!")
    print(f"Tôi đã chọn một số ngẫu nhiên giữa {n1} và {n2}. Hãy đoán xem đó là số nào!")

    while True:
        # Người chơi nhập số dự đoán
        player_guess = int(input("Nhập số bạn đoán: "))
        attempts += 1

        # Kiểm tra kết quả
        if player_guess < number_to_guess:
            print("Số bạn đoán nhỏ hơn số tôi chọn. Hãy thử lại!")
        elif player_guess > number_to_guess:
            print("Số bạn đoán lớn hơn số tôi chọn. Hãy thử lại!")
        else:
            print(f"Chúc mừng bạn! Bạn đã đoán đúng số {number_to_guess} sau {attempts} lần thử!")
            break

# Gọi hàm để chơi trò chơi
guess_the_number()
```

# #Trò chơi "Rock, Paper, Scissors" (Búa, Giấy, Kéo) giữa người chơi và máy tính.
Trò chơi này sử dụng thư viện random để máy tính chọn một trong ba tùy chọn: Bao, Búa hoặc Kéo. Người chơi sẽ nhập lựa chọn của mình và kết quả sẽ được quyết định theo quy tắc của trò chơi.
Trò chơi Búa, Giấy, Kéo là một trò chơi đơn giản nhưng thú vị, nơi người chơi sẽ đối đầu với máy tính bằng cách chọn một trong ba lựa chọn: Búa (rock), Giấy (paper) hoặc Kéo (scissors). Mỗi lựa chọn có một mối quan hệ chiến thắng hoặc thua với các lựa chọn còn lại. Mục tiêu của trò chơi là chọn một vật mà "đánh bại" vật mà máy tính chọn.

# Mục tiêu của trò chơi:
Đoán lựa chọn của đối thủ (máy tính) và chọn một vật sao cho lựa chọn của bạn chiến thắng.

Trò chơi kết thúc khi bạn và máy tính có kết quả hoặc hòa hoặc bạn chiến thắng.

# Luật chơi:
Búa (Rock) đánh bại Kéo (Scissors) (Búa đập Kéo).

Giấy (Paper) đánh bại Búa (Rock) (Giấy bao Búa).

Kéo (Scissors) đánh bại Giấy (Paper) (Kéo cắt Giấy).

Nếu cả người chơi và máy tính đều chọn cùng một vật, kết quả là hòa.

# Hướng dẫn cách chơi:
Bắt đầu trò chơi:

Trò chơi yêu cầu bạn nhập một lựa chọn từ ba lựa chọn: rock, paper, hoặc scissors (Búa, Giấy, Kéo).

Máy tính chọn ngẫu nhiên:

Sau khi bạn nhập lựa chọn của mình, máy tính sẽ chọn ngẫu nhiên một trong ba vật: rock, paper, hoặc scissors.

# So sánh kết quả:

Trò chơi sẽ so sánh lựa chọn của bạn với lựa chọn của máy tính:

Nếu bạn chọn "Búa" và máy tính chọn "Kéo", bạn thắng.

Nếu bạn chọn "Giấy" và máy tính chọn "Búa", bạn thắng.

Nếu bạn chọn "Kéo" và máy tính chọn "Giấy", bạn thắng.

Nếu bạn và máy tính chọn cùng một vật, kết quả là hòa.

# Kết thúc trò chơi:

Sau mỗi lượt, trò chơi sẽ thông báo kết quả: bạn thắng, máy tính thắng, hoặc hòa.

Bạn có thể chơi lại bằng cách nhập lại lựa chọn của mình.

# Ví dụ về cách chơi:
# Lượt 1:

Bạn nhập "rock".

Máy tính chọn "scissors".

Kết quả: Bạn thắng! (Búa đập Kéo)

# Lượt 2:

Bạn nhập "paper".

Máy tính chọn "rock".

Kết quả: Bạn thắng! (Giấy bao Búa)

# Lượt 3:

Bạn nhập "scissors".

Máy tính chọn "scissors".

Kết quả: Hòa!



```python
import random

def rock_paper_scissors():
    # Danh sách các tùy chọn: Giấy, Búa, Kéo
    choices = ["giấy", "búa", "kéo"]

    print("Chào mừng bạn đến với trò chơi Giấy (Bao), Búa, Kéo!")
    print("Chọn một trong ba tùy chọn: giấy, búa, kéo")

    while True:
        # Người chơi nhập lựa chọn
        player_choice = input(" Nhập lựa chọn của bạn (giấy, búa, kéo) hoặc 'thoát hoặc t' để dừng chơi: ").lower()

        # Kiểm tra nếu người chơi muốn thoát
        if player_choice == 'thoát' or player_choice == 't' :
            print("Cảm ơn bạn đã chơi!")
            break

        # Kiểm tra nếu người chơi nhập lựa chọn hợp lệ
        if player_choice not in choices:
            print("Lựa chọn không hợp lệ! Vui lòng nhập 'bao', 'búa' hoặc 'kéo'.")
            continue

        # Máy tính chọn ngẫu nhiên
        computer_choice = random.choice(choices)
        print(f"Máy tính chọn: {computer_choice}")

        # Quy tắc quyết định thắng thua
        if player_choice == computer_choice:
            print("Hòa! Cả hai đều chọn", player_choice)
        elif (player_choice == "búa" and computer_choice == "kéo") or \
             (player_choice == "giấy" and computer_choice == "búa") or \
             (player_choice == "kéo" and computer_choice == "giấy"):
            print(f"Chúc mừng bạn! Bạn thắng, {player_choice} đánh bại {computer_choice}.")
        else:
            print(f"Rất tiếc, bạn thua! {computer_choice} đánh bại {player_choice}.")

# Gọi hàm để chơi trò chơi
rock_paper_scissors()
```

Update
Yêu cầu nhập vào số lần chơi phải là số lẻ.

Kết thúc trò chơi và in ra số lượt thắng, hòa của mỗi bên (người chơi và máy tính).

Bên thắng chung cuộc (người chơi hoặc máy tính) cũng sẽ được hiển thị.

```python
import random

def rock_paper_scissors():
    # Danh sách các tùy chọn: Giấy, Búa, Kéo
    choices = ["g", "b", "k"]
    choices_full = {"g": "giấy", "b": "búa", "k": "kéo"}

    # Yêu cầu người chơi nhập số lần chơi, phải là số lẻ
    while True:
        try:
            rounds = int(input("Nhập số lần chơi (phải là số lẻ): "))
            if rounds % 2 == 1:  # Kiểm tra nếu số lần chơi là số lẻ
                break
            else:
                print("Số lần chơi phải là số lẻ. Vui lòng nhập lại.")
        except ValueError:
            print("Vui lòng nhập một số hợp lệ.")

    # Biến để lưu số lượt thắng của người chơi, máy tính và số lượt hòa
    player_wins = 0
    computer_wins = 0
    draws = 0

    print("Chào mừng bạn đến với trò chơi Giấy (Bao), Búa, Kéo!")
    print("Chọn một trong ba tùy chọn: 'g' cho giấy, 'b' cho búa, 'k' cho kéo")

    for i in range(rounds):
        print(f"\n--- Vòng chơi {i+1} ---")

        # Người chơi nhập lựa chọn
        player_choice = input("Nhập lựa chọn của bạn (g - giấy, b - búa, k - kéo) hoặc 'thoát' hoặc 't' để dừng chơi: ").lower()

        # Kiểm tra nếu người chơi muốn thoát
        if player_choice == 'thoát' or player_choice == 't':
            print("Cảm ơn bạn đã chơi!")
            break

        # Kiểm tra nếu người chơi nhập lựa chọn hợp lệ
        if player_choice not in choices:
            print("Lựa chọn không hợp lệ! Vui lòng nhập 'g' cho giấy, 'b' cho búa hoặc 'k' cho kéo.")
            continue

        # Máy tính chọn ngẫu nhiên
        computer_choice = random.choice(choices)
        print(f"Máy tính chọn: {choices_full[computer_choice]}")

        # Quy tắc quyết định thắng thua
        if player_choice == computer_choice:
            print(f"Hòa! Cả hai đều chọn {choices_full[player_choice]}")
            draws += 1
        elif (player_choice == "b" and computer_choice == "k") or \
             (player_choice == "g" and computer_choice == "b") or \
             (player_choice == "k" and computer_choice == "g"):
            print(f"Chúc mừng bạn! Bạn thắng, {choices_full[player_choice]} đánh bại {choices_full[computer_choice]}.")
            player_wins += 1
        else:
            print(f"Rất tiếc, bạn thua! {choices_full[computer_choice]} đánh bại {choices_full[player_choice]}.")
            computer_wins += 1

    # Sau khi kết thúc tất cả các vòng, in kết quả chung cuộc
    print("\n--- Kết quả ---")
    print(f"Số lượt thắng của bạn: {player_wins}")
    print(f"Số lượt thắng của máy tính: {computer_wins}")
    print(f"Số lượt hòa: {draws}")

    # Xác định bên thắng chung cuộc
    if player_wins > computer_wins:
        print("Chúc mừng! Bạn là người thắng cuộc!")
    elif player_wins < computer_wins:
        print("Máy tính thắng cuộc!")
    else:
        print("Trận đấu hòa! Cả hai đều có số lượt thắng như nhau.")

# Gọi hàm để chơi trò chơi
rock_paper_scissors()
```

Trò chơi Hangman
Luật chơi:
Từ cần đoán: Máy tính chọn một từ ngẫu nhiên từ một danh sách đã cho trước.

Dấu gạch dưới: Từ cần đoán sẽ được hiển thị dưới dạng các dấu gạch dưới, mỗi dấu gạch đại diện cho một chữ cái của từ.

Đoán chữ: Người chơi sẽ nhập một chữ cái mỗi lần. Nếu chữ cái đúng, nó sẽ được thay thế vào vị trí tương ứng trong từ.

Lượt đoán sai: Nếu người chơi đoán sai một chữ cái, một phần của hình người treo sẽ xuất hiện (từ đầu chỉ có đầu, rồi thêm thân, tay, chân...).

Kết thúc trò chơi:
Thắng: Người chơi đoán đúng toàn bộ từ.

Thua: Máy tính hoàn tất hình người treo sau số lượt đoán sai tối đa (thường là 6 lần).

Hướng dẫn cách chơi: Bắt đầu trò chơi:

Trò chơi bắt đầu khi máy tính chọn một từ ngẫu nhiên từ một danh sách.

Từ sẽ được hiển thị dưới dạng các dấu gạch dưới, ví dụ: _ _ _ _ _.

Người chơi đoán chữ:

Người chơi nhập một chữ cái vào mỗi lần.

Nếu chữ cái đúng, chữ cái đó sẽ được hiển thị tại vị trí tương ứng trong từ.

Nếu chữ cái sai, hình người treo sẽ được vẽ thêm một phần.

Lượt đoán:
Người chơi có một số lượt đoán sai (thường là 6 lượt).

Khi số lượt đoán sai hết, trò chơi kết thúc với kết quả thua.

Chọn từ ngẫu nhiên:
Máy tính sẽ chọn ngẫu nhiên một từ từ danh sách và yêu cầu người chơi đoán.

Hình người treo:

Hình người treo sẽ vẽ dần dần theo số lần đoán sai của người chơi.

Kết thúc trò chơi:
Nếu người chơi đoán đúng trước khi hết lượt, trò chơi kết thúc với chiến thắng.

Nếu hết lượt và không đoán đúng từ, người chơi thua cuộc.

Ví dụ chơi:

Người chơi đoán "t", hệ thống sẽ cập nhật: p y t _ _ _

Người chơi đoán "z", hệ thống sẽ thông báo: "Chữ cái 'z' không có trong từ. Bạn còn 5 lượt thử."

Tiếp tục cho đến khi đoán đúng hết từ hoặc hết lượt đoán.

Thua: Nếu người chơi đoán sai hết 6 lượt, trò chơi sẽ thông báo thua cuộc.

```python
import random
import time

# Danh sách các từ cho trò chơi
word_list = ["python", "java", "kaggle", "programming", "machine", "data", "science", "algorithm"]
# Chọn một từ ngẫu nhiên từ danh sách
word_to_guess = random.choice(word_list)

# Tạo một danh sách dấu gạch dưới tương ứng với số lượng ký tự trong từ cần đoán
guessed_word = ["_"] * len(word_to_guess)
attempts_left = 6  # Số lần thử tối đa (tương ứng với số lần treo hình)
guessed_letters = []  # Danh sách các chữ cái đã đoán

# Hàm vẽ hình người treo
def hangman(attempts):
    stages = [
        '''
           ------
           |    |
           |
           |
           |
           |
        ------
        ''',
        '''
           ------
           |    |
           |    O
           |
           |
           |
        ------
        ''',
        '''
           ------
           |    |
           |    O
           |    |
           |
           |
        ------
        ''',
        '''
           ------
           |    |
           |    O
           |   /|
           |
           |
        ------
        ''',
        '''
           ------
           |    |
           |    O
           |   /|\\
           |
           |
        ------
        ''',
        '''
           ------
           |    |
           |    O
           |   /|\\
           |   /
           |
        ------
        ''',
        '''
           ------
           |    |
           |    O
           |   /|\\
           |   / \\
           |
        ------
        '''
    ]
    return stages[attempts]

# Hàm kiểm tra xem người chơi có đoán đúng từ hay không
def display_word():
    print(" ".join(guessed_word))

# Hàm bắt đầu trò chơi
def play_hangman():
    global attempts_left, guessed_word, guessed_letters, word_to_guess
    print("Chào mừng bạn đến với trò chơi Hangman!")
    print("Bạn có 6 lần thử để đoán từ. Hãy bắt đầu!")
    time.sleep(1)

    while attempts_left > 0:
        print(hangman(6 - attempts_left))  # In hình treo tương ứng với số lần thử còn lại
        display_word()  # Hiển thị từ cần đoán với các dấu gạch dưới
        print(f"Số lần thử còn lại: {attempts_left}")

        guess = input("Nhập chữ cái bạn đoán: ").lower()  # Lấy chữ cái người chơi nhập vào

        # Kiểm tra nếu người chơi đã đoán chữ cái này rồi
        if guess in guessed_letters:
            print(f"Bạn đã đoán chữ '{guess}' rồi. Hãy thử chữ khác.")
            continue

        # Kiểm tra nếu người chơi nhập vào một ký tự hợp lệ
        if len(guess) != 1 or not guess.isalpha():
            print("Vui lòng nhập chỉ một chữ cái hợp lệ.")
            continue

        guessed_letters.append(guess)  # Thêm chữ cái vào danh sách đã đoán

        # Kiểm tra nếu chữ cái đoán đúng
        if guess in word_to_guess:
            print(f"Chúc mừng! Chữ cái '{guess}' đúng.")
            for i in range(len(word_to_guess)):
                if word_to_guess[i] == guess:
                    guessed_word[i] = guess
        else:
            attempts_left -= 1
            print(f"Rất tiếc, chữ cái '{guess}' không có trong từ.")

        # Kiểm tra nếu người chơi đã đoán đúng toàn bộ từ
        if "_" not in guessed_word:
            print(f"Chúc mừng! Bạn đã đoán đúng từ '{word_to_guess}'")
            break

    if attempts_left == 0:
        print(hangman(6 - attempts_left))
        print(f"Bạn đã hết lượt thử. Từ cần đoán là '{word_to_guess}'. Chúc bạn may mắn lần sau!")

# Chạy trò chơi
play_hangman()
```
